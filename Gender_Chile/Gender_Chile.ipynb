```python
import pandas as pd

df = pd.read_excel(r"df10.xls")
```

# How many males are technicians & directors and females are technicians & directors

```python
sex_occ = df.groupby('SEXO')['OCUPACION'].value_counts()
sex_occ = pd.DataFrame(sex_occ)
sex_occ
```

# Check median age to become technician and director for males and females

```python
age_occ = df.groupby(['SEXO','OCUPACION'])['EDAD'].median()
age_occ  = pd.DataFrame(age_occ)
age_occ
```

# Check ranking given by employeer to male and female employees based on the occupation - checking the mean


```python
f = {'EV TOTAL':'mean','TOTAL_SUP':'mean','TOTAL_PEER':'mean',
    'TOTAL_SUB':'mean'}
rank_occ = df.groupby(['SEXO','OCUPACION']).agg(f)
rank_occ  = pd.DataFrame(rank_occ)
rank_occ
```

# Check ranking given by employeer to male and female employees based on the occupation - checking the median

```python
#ranking given by employeer to male and female employees based on the occupation - checking the median
f1 = {'EV TOTAL':'median','TOTAL_SUP':'median','TOTAL_PEER':'median','TOTAL_SUB':'median'}
rank1_occ = df.groupby(['SEXO','OCUPACION']).agg(f1)
rank1_occ  = pd.DataFrame(rank1_occ)
rank1_occ
```

# Comparing EV total, Total_Sup, Total_Peer, Total_Sub amongst males and females for the two roles (technician & director) shows us that they are similar for both the sexes 

```python
col =['EV TOTAL','TOTAL_SUP','TOTAL_PEER','TOTAL_SUB']

import seaborn as sns
for it in col:

   sns.catplot(x="OCUPACION", y=it, hue="SEXO", kind="box", data=df)
```

# Wordcloud for females

```python
#gen
#dataframe for females
fem = df[df['SEXO']=='MUJER']
fem.head(5)

text = " ".join(cat.split()[1] for cat in fem['FORTALEZAS'])
import matplotlib.pyplot as plt

from wordcloud import WordCloud

wordcloud = WordCloud(collocations = False, background_color = 'white').generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

# Some common descriptions for females

```python
import itertools
import collections
import nltk
import string
import re

text = re.sub(r'[^\w\s]','',text)
txt = nltk.word_tokenize(text)


# Create counter
counts_no_urls = collections.Counter(txt)

counts_no_urls.most_common(15)
```
